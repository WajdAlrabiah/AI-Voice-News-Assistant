##**The Lab**

Behind the scenes

```python
# Step 2: Task 1 - Translate Text from English to Arabic
def translate_text(text):
    translator = Translator()
    translation = translator.translate(text, src='en', dest='ar')
    print(f"Translated text: {translation.text}")
```

```python
# Step 3: Task 2 - Retrieve Headline News Using the News API
def get_headline_news(NEWS_API_KEY):
    url = f'https://newsapi.org/v2/top-headlines?country=us&apiKey={NEWS_API_KEY}'
    response = requests.get(url)
    headlines = response.json().get('articles', [])

    print("Headlines:")
    for article in headlines:
        print(article['title'])
```

```python
# Step 4: Task 3 - Retrieve Flight Details Using the Aviation Stack API
def get_flight_details(AVIATION_API_KEY, flight_number):
    url = f"https://api.aviationstack.com/v1/flights"
    params = {
        'access_key': AVIATION_API_KEY,
        'flight_iata': flight_number
    }

    # Making the API request
    response = requests.get(url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        if data['data']:
            flight_info = data['data'][0]
            details = {
                'Airline': flight_info.get('airline', {}).get('name', 'N/A'),
                'Departure Airport': flight_info.get('departure', {}).get('airport', 'N/A'),
                'Arrival Airport': flight_info.get('arrival', {}).get('airport', 'N/A'),
                'Departure Time': flight_info.get('departure', {}).get('estimated', 'N/A'),
                'Arrival Time': flight_info.get('arrival', {}).get('estimated', 'N/A'),
                'Status': flight_info.get('flight_status', 'N/A')
            }

            # Displaying the details in a readable format
            print("\nFlight Details:")
            for key, value in details.items():
                print(f"{key}: {value}")
            return details
        else:
            print("No flight information found for this flight number.")
            return {}
    else:
        print(f"Error: Unable to retrieve flight details. Status code: {response.status_code}")
        return {}
```

```python
import os
import time
from gtts import gTTS
from IPython.display import Audio, display

def read_headlines(headlines):
    for i, headline in enumerate(headlines):
        print(f"Playing headline {i+1}: {headline}")

        # Convert headline text to speech
        tts = gTTS(text=headline, lang='en')
        filename = f"headline_{i}.mp3"
        tts.save(filename)

        # Play the saved audio file
        display(Audio(filename, autoplay=True))

        # Wait for a few seconds to allow audio to play
        time.sleep(5)  # Adjust if headline length varies

        # Remove the audio file after playing to keep the directory clean
        if os.path.exists(filename):
            os.remove(filename)
```

## **Interface**

```python
# Step 6: Main Function to Ask User for Task
def main():
    print("Welcome to the AI News Translator!")
    while True:
        print("\nChoose a task:")
        print("1. Translate text from English to Arabic")
        print("2. Retrieve headline news")
        print("3. Retrieve flight details")
        print("4. Read headlines aloud")
        print("5. Exit")

        choice = input("Enter the number of the task you want to perform: ")

        if choice == '1':
            text_to_translate = input("Enter text to translate: ")
            translate_text(text_to_translate)
        elif choice == '2':
            get_headline_news(NEWS_API_KEY)
        elif choice == '3':
            flight_number = input("Enter the flight number: ")
            get_flight_details(AVIATION_API_KEY, flight_number)
        elif choice == '4':
            read_headlines()
        elif choice == '5':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

# Step 7: Run the Main Function
if __name__ == "__main__":
    main()
```
